```python
import pandas as pd
import io
import matplotlib.pyplot as plt
import subprocess
%matplotlib inline
```

```python
def profile():
    csv = """idx, size , main_ins  , main_qsort, keys_ins  , keys_qsort
0  , 50   , 66937     , 22655     , 38755     , 12881     
1  , 100  , 234494    , 51294     , 139686    , 35948     
2  , 300  , 1976292   , 189788    , 1326587   , 119717    
3  , 500  , 5294935   , 366855    , 3885049   , 232534    
4  , 1000 , 21258106  , 894144    , -25462350 , 642854    
"""
    df = pd.read_csv(io.StringIO(csv))
    df.columns = df.columns.str.replace(' ', '')
    return df
```

```python
def plot_profile(df, log_x=True, log_y=True):
    plt.figure(figsize=(8, 6), dpi=80)

    plt.plot(df["size"], df["main_qsort"], label="main qsort")
    plt.plot(df["size"], df["main_ins"], label="main insertion")
    plt.plot(df["size"], df["keys_qsort"], label="keys qsort")
    plt.plot(df["size"], df["keys_ins"], label="keys insertion")


    plt.yscale("log" if log_y else "linear")
    plt.xscale("log" if log_x else "linear")
    plt.legend()
    plt.xlabel("# elements")
    plt.ylabel("$\\mu s$")
    plt.plot(figsize=(10, 10))
```

```python
df = profile()
df
```

```python
plot_profile(df, log_x=False, log_y=False)
```
