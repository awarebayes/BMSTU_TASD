```python
import pandas as pd
import io
import matplotlib.pyplot as plt
import subprocess
%matplotlib inline
```

```python
def profile():
    csv = """
idx, size , main_ins  , main_qsort, keys_ins  , keys_qsort
0  , 50   , 74997     , 14488     , 46499     , 14742     
1  , 100  , 154738    , 29819     , 157322    , 33038     
2  , 300  , 993315    , 107596    , 1448529   , 127228    
3  , 500  , 2669218   , 192401    , 3755676   , 231264    
4  , 1000 , 10235822  , 428684    , 14836019  , 516590
"""
    df = pd.read_csv(io.StringIO(csv))
    df.columns = df.columns.str.replace(' ', '')
    return df
```

```python
df = profile()
plt.figure(figsize=(8, 6), dpi=80)
plt.plot(df["size"], df["main_qsort"], label="main qsort")
plt.plot(df["size"], df["main_ins"], label="main insertion")
plt.plot(df["size"], df["keys_qsort"], label="keys qsort")
plt.plot(df["size"], df["keys_ins"], label="keys insertion")
plt.yscale("linear")
plt.xscale("linear")
plt.legend()
plt.xlabel("# elements")
plt.ylabel("$\\mu s$")
plt.plot(figsize=(10, 10))
```
