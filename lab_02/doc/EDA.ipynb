```python
import pandas as pd
import io
import matplotlib.pyplot as plt
import subprocess
%matplotlib inline
```

```python
def profile():
    csv = """
idx, size , main_ins  , main_qsort, keys_ins  , keys_qsort
0  , 50   , 99236     , 12176     , 39637     , 3894      
1  , 100  , 393030     , 25298     , 136272    , 8657      
2  , 300  , 644668    , 89902     , 1113130   , 31188     
3  , 500  , 1715942   , 157273    , 157965   , 56441     
4  , 1000 , 6836234   , 351973    , 2523300  , 132131     
"""
    df = pd.read_csv(io.StringIO(csv))
    df.columns = df.columns.str.replace(' ', '')
    return df
```

```python
df = profile()
plt.figure(figsize=(8, 6), dpi=80)
plt.plot(df["size"], df["main_qsort"], label="main qsort")
plt.plot(df["size"], df["main_ins"], label="main insertion")
plt.plot(df["size"], df["keys_qsort"], label="keys qsort")
plt.plot(df["size"], df["keys_ins"], label="keys insertion")
plt.yscale("linear")
plt.xscale("linear")
plt.legend()
plt.xlabel("# elements")
plt.ylabel("$\\mu s$")
plt.plot(figsize=(10, 10))
```
