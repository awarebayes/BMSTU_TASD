```python
import pandas as pd
import io
import matplotlib.pyplot as plt
import subprocess
%matplotlib inline
```

```python
def profile():
    csv = """idx, size , main_ins  , main_qsort, keys_ins  , keys_qsort
0  , 50   , 430015    , 14713     , 42469     , 3463      
1  , 100  , 294219    , 36902     , 148711    , 8595      
2  , 300  , 1264344   , 102430    , 1212284   , 32030     
3  , 500  , 2950718   , 185493    , 3309855   , 56744     
4  , 1000 , 10420676  , 436738    , 13284580  , 127525    
"""
    df = pd.read_csv(io.StringIO(csv))
    df.columns = df.columns.str.replace(' ', '')
    return df
```

```python
df = profile()
plt.figure(figsize=(8, 6), dpi=80)
plt.plot(df["size"], df["main_qsort"], label="main qsort")
plt.plot(df["size"], df["main_ins"], label="main insertion")
plt.plot(df["size"], df["keys_qsort"], label="keys qsort")
plt.plot(df["size"], df["keys_ins"], label="keys insertion")
plt.yscale("linear")
plt.xscale("linear")
plt.legend()
plt.xlabel("# elements")
plt.ylabel("$\\mu s$")
plt.plot(figsize=(10, 10))
```
