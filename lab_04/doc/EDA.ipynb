```python
import pandas as pd
import io
import matplotlib.pyplot as plt
import subprocess
%matplotlib inline
```

```python
def profile():
    csv = """size,     stack_r,            list_r,           stack_p,           list_p
                 10,             505,             531,             537,             773,
                 50,            1289,            3067,            1659,            4069,
                100,            1901,            6049,            3269,            7998,
                500,            7525,           30189,            9562,           19704,
               1000,            7175,           30142,           13615,           42131,
               5000,           34797,          153404,           35692,          151552,
              10000,           69644,          309015,           72181,          308703,
"""
    df = pd.read_csv(io.StringIO(csv), index_col=False)
    df.columns = df.columns.str.replace(' ', '')
    return df
```

```python
df.index
```

```python
df = profile()
plt.figure(figsize=(8, 6), dpi=80)
plt.plot(df["size"], df["stack_r"], label="stack_random")
plt.plot(df["size"], df["stack_p"], label="stack_palindrome")
plt.plot(df["size"], df["list_r"], label="list_random")
plt.plot(df["size"], df["list_p"], label="list_palindrome")
plt.yscale("linear")
plt.xscale("log")
plt.legend()
plt.title("Is palindrome")
plt.xlabel("# elements")
plt.ylabel("ticks")
plt.plot(figsize=(10, 10))
```
